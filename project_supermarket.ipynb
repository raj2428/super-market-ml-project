```python
# part 2 questions

sns.countplot(x='Gender',data=df)
plt.title("Costomer Gender Distribution")
plt.show()

df['Gender'].value_counts()
```

```python
sns.histplot(df['Rating'], kde=True, bins=20)
plt.title("Customer Rating Distribution")
plt.show()

print("Skewness:", df['Rating'].skew())
```

```python
branch_sales = df.groupby("Branch")["Total"].sum()
branch_sales.plot(kind="bar", title="Aggregate Sales per Branch")
plt.show()

branch_sales
```

```python
sns.countplot(x="Payment", data=df)
plt.title("Most Popular Payment Method")
plt.show()

df['Payment'].value_counts()
```

```python
sns.countplot(x="Branch", hue="Payment", data=df)
plt.title("Branch-wise Payment Distribution")
plt.show()
```

```python
sns.scatterplot(x="gross income", y="Rating", data=df, alpha=0.6)
plt.title("gross Income vs Rating")
plt.show()

```

```python
sns.boxplot(x="Branch", y="Rating", data=df)
plt.title("Rating Distribution Across Branches")
plt.show()
```

```python
branch_profit = df.groupby("Branch")["gross income"].sum()
branch_profit.plot(kind="bar", title="Profit by Branch")
plt.show()

branch_profit
```

```python
sns.boxplot(x="Gender", y="gross income", data=df)
plt.title("Gender vs gross Income")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
daily_income = df.groupby("Date")["gross income"].sum()
daily_income.plot(figsize=(10,5), title="gross Income Over Time")
plt.show()
```

```python
product_income = df.groupby("Product line")["gross income"].sum().sort_values(ascending=False)
product_income.plot(kind="bar", title="Product Line Income")
plt.show()

product_income
```

```python
sns.barplot(x="Product line", y="Total", hue="Gender", data=df, estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Male vs Female Spending Pattern")
plt.show()
```

```python
df["Quantity"].sum()
```

```python
df["Day"] = df["Date"].dt.day_name()
day_sales = df.groupby("Day")["Total"].sum().sort_values(ascending=False)
day_sales.plot(kind="bar", title="Sales by Day of Week")
plt.show()

day_sales
```

```python
df["Hour"] = pd.to_datetime(df["Time"]).dt.hour
hourly_sales = df.groupby("Hour")["Total"].sum()
hourly_sales.plot(kind="bar", title="Busiest Hour of Day")
plt.show()

hourly_sales
```

```python
best_line = df.groupby("Product line")["Total"].sum().sort_values(ascending=False)
best_line
```

```python
city_income = df.groupby("City")["gross income"].sum().sort_values(ascending=False)
sns.barplot(x=city_income.index, y=city_income.values)
plt.title("City-wise Gross Income")
plt.show()

city_income
```
